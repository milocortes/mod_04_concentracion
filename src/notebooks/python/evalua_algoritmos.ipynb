## Función de prueba: Himmelblau

```python
# Función Himmelblau 
def himmelblau(X):
    x,y = X
    return (x**2 + y -11)**2 + (x + y**2 -7)**2
```

```python
# Ejecutamos el algoritmo genético binario

m = 100
n_variables = 2
i_sup_vec = [5, 5]
i_inf_vec = [-5, -5]
precision = 10
maxiter = 60
prob_c = 0.9

fitness_genetico, x_best_genetico = genetico_binario(himmelblau,m,n_variables,i_sup_vec,i_inf_vec,precision,maxiter,prob_c)
x_best_genetico
```

```python
# Ejecutamos el algoritmo DE
n_dim = 2               
lb = np.array([-5, -5]) 
ub = np.array([5, 5] )

maxiter = 60  

pc = 0.9 
pop_size = 100   


fitness_de, x_best_de = DE(himmelblau,pop_size,maxiter,pc,lb,ub,step_size = 0.8)

x_best_de
```

```python
# Tamaño de la población
n = 100
# Número de variables
n_var = 2
l_bounds = np.array([-5,-5])
u_bounds = np.array([5,5])
maxiter = 60
# Social scaling parameter
α = 0.8
# Cognitive scaling parameter
β = 0.8
# velocity inertia
w = 0.5

fitness_pso, x_best_pso = PSO(himmelblau, n, maxiter, n_var, l_bounds, u_bounds, α, β, w)
x_best_pso
```

```python
# Comparamos el fitness de los algoritmos
import matplotlib.pyplot as plt 

plt.plot(range(maxiter), fitness_de, label ="DE")
plt.plot(range(maxiter), fitness_pso, label ="PSO")
plt.plot(range(maxiter), fitness_genetico, label ="Genético")

plt.legend()
plt.title("Comparación de algoritmos de optimización")
```

```python
min_value = 100000
x_best = None
algo_best = None

for algo,x in zip(["Genético","DE","PSO"],[x_best_genetico,x_best_de,x_best_pso]):
    if himmelblau(x) < min_value:
        min_value = himmelblau(x)
        x_best = x
        algo_best = algo
print(f"El valor mínimo de la función es {min_value}\nEl mejor vector fue el del algoritmo {algo_best}\nx_best: {x_best}")
```
